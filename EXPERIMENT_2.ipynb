```python
#1
print("----------------------------------------")
x = 15
y = 77

addition = x + y
multiplication = x * y

division = x / y

subtraction = x - y

print("Addition (x + y):", addition)
print("Multiplication (x * y):", multiplication)
print("Division (x / y):", division)
print("Subtraction (x - y):", subtraction)

#2
print("----------------------------------------")
import math
radius = float(input("Enter the radius of the circle: "))
area = math.pi * radius ** 2
print("The area of the circle is:", area)

#3
print("-------------------------------------------")
x = 4
y = 3
result = (x + y) * (x + y)
print("The result of (x + y) * (x + y) is:", result)

#4
print("---------------------------------------------")
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = math.sqrt(a*2 + b*2)
print("The length of the hypotenuse is:", c)

#5
print("------------------------------------------------")
P = float(input("Enter the principal amount: "))
R = float(input("Enter the rate of interest (in %): "))
T = float(input("Enter the time period (in years): "))
SI = (P * R * T) / 100
print("The Simple Interest is:", SI)

#6
print("--------------------------------------------------")
a = float(input("Enter the length of side a: "))
b = float(input("Enter the length of side b: "))
c = float(input("Enter the length of side c: "))
s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("The area of the triangle is:", area)

#7
print("-------------------------------------------------")
total_seconds = int(input("Enter the number of seconds: "))
hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
remaining_seconds = total_seconds % 60
print(f"{total_seconds} seconds is equivalent to {hours} hours, {minutes} minutes, and {remaining_seconds} seconds.")

#8
print("---------------------------------------------------")
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
a = a + b
b = a - b
a = a - b
print("After swapping:")
print("a =", a)
print("b =", b)

#9
print("---------------------------------------------------")
n = int(input("Enter the value of n: "))
sum_n = n * (n + 1) // 2
print("The sum of the first", n, "natural numbers is:", sum_n)

#10
print("--------------------------------------------------------")
print("A B A&B A|B A^B")
for A in [0, 1]:
    for B in [0, 1]:
        print(A, B, A & B, A | B, A ^ B)

#11
print("---------------------------------------------------------")
num = int(input("Enter a number: "))
left_shift = num << 2  # Shift left by 2 positions
right_shift = num >> 2  # Shift right by 2 positions
print(f"Left shift of {num} by 2 is: {left_shift}")
print(f"Right shift of {num} by 2 is: {right_shift}")

#12
print("--------------------------------------------------------")
sequence = (10, 20, 56, 78, 89)
num = int(input("Enter a number to check: "))
if num in sequence:
    print(f"{num} is in the sequence.")
else:
    print(f"{num} is not in the sequence.")

#13
print("-------------------------------------------------------------")
string = "Hello, World!"
char = input("Enter a character to check: ")
if char in string:
    print(f"'{char}' is in the string.")
else:
    print(f"'{char}' is not in the string.")
```
